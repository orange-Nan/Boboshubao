```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#原始数据
months = np.arange(1,13)  #原始数据（散点）
tem = [1.43,3.91,10.98,16.75,22.29,25.29,28.75,27.55,23.16,17.70,np.nan,6.59]  #11月没有数据
```

```python
# 设置字体格式（Times New Roman）
plt.rc('font', family='Times New Roman')
plt.rcParams['font.weight'] = 'normal'
plt.rcParams['axes.labelweight'] = 'normal'
plt.rcParams['axes.unicode_minus'] = True
```

```python
#数据分段
months1 = np.arange(1,11) #1到10月（实线）
tem1 = [1.43,3.91,10.98,16.75,22.29,25.29,28.75,27.55,23.16,17.70]
months2 = [10,12]  #10到12月（虚线）
tem2 = [17.70,6.59]

fig = plt.figure(figsize=(6,6),dpi=200)
ax1 = fig.add_subplot(111)
ax1.plot(months1,tem1,c='m')  #1到10月（实线）
ax1.plot(months2,tem2,ls='--',c='m')  #10到12月（虚线）
ax1.scatter(months,tem,c='m')  #原始数据（散点）
ax1.tick_params(axis='x',which='major',direction='out',length=5,width=1,labelsize=15)
ax1.tick_params(axis='y',which='major',direction='out',length=5,width=1,labelsize=15)
ax1.set_xticks(months)  #设置x轴刻度
ax1.set_yticks(np.arange(0,40,5))
ax1.set_xlabel('Time(month)',fontsize=20,labelpad=10)
ax1.set_ylabel('Temperature(°C)',fontsize=20,labelpad=10)

#保存图片
plt.savefig('C:/Users/LULU/Desktop/fig4.png',dpi=200,bbox_inches = 'tight')
plt.show()
```
