## 【Part 1】绘制tem00-19.csv时间序列图

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import MultipleLocator
```

```python
df = pd.read_csv('D:/寒假Python课程/Lesson 0/data/tem00-19.csv')
```

```python
# 设置字体格式（Times New Roman）
plt.rc('font', family='Times New Roman')
plt.rcParams['font.weight'] = 'normal'
plt.rcParams['axes.labelweight'] = 'normal'
plt.rcParams['axes.unicode_minus'] = True
```

```python
fig = plt.figure(figsize=(20,5),dpi=600)
ax1 = fig.add_subplot(111)
ax1.plot(df.index,df['tem'],c='k')
ax1.set_yticks(np.arange(0,40,5))
ax1.xaxis.set_major_locator(MultipleLocator(12))
ax1.set_xticklabels(np.arange(1999,2021,1))
ax1.tick_params(axis='x',which='major',direction='in',length=8,width=1,labelsize=18)
ax1.tick_params(axis='y',which='major',direction='in',length=8,width=1,labelsize=18)
ax1.set_xlabel('time(year)',fontsize=25)
ax1.set_title('2000-2019 Temperature(°C)',fontsize=30,y=1.05)
plt.savefig('C:/Users/LULU/Desktop/fig2.png',dpi=200)  #图片保存路径
plt.show()
```

## 【Part 2】绘制tem20.csv逐月气温柱状图

```python
df1 = pd.read_csv('D:/寒假Python课程/Lesson 0/data/tem20.csv')
```

```python
x = np.arange(1,13,1)  # the label locations
width = 0.8  # the width of the bars

fig = plt.figure(figsize=(6,6),dpi=600)
ax1 = fig.add_subplot(111)
rects1 = ax1.bar(x, df1['tem'], width)
ax1.bar_label(rects1, fmt='%.2f', padding=3)

ax1.set_yticks(np.arange(0,40,5))
ax1.xaxis.set_major_locator(MultipleLocator(1))
ax1.tick_params(axis='x',which='major',direction='in',length=8,width=1,labelsize=18)
ax1.tick_params(axis='y',which='major',direction='in',length=8,width=1,labelsize=18)
ax1.set_xlabel('time(month)',fontsize=25)
ax1.set_title('2020 Temperature(°C)',fontsize=30,y=1.05)
plt.savefig('C:/Users/LULU/Desktop/fig4.png',dpi=200)  #图片保存路径
plt.show()
```
